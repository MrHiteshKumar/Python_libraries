```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# import datasets
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
Categorical Plots

Categorical Scatter Plot
 - Stripplot
 - Swarmplot
Categorical Distribution Plots
 - Boxplot
 - Violinplot
Categorical Estimate Plot -> for central tendency
 - Barplot
 - Pointplot
 - Countplot
Figure level function -> catplot
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
# strip plot -> axes level function
sns.stripplot(data=tips, x='day', y='total_bill', jitter=False)
```

```python
# catplot ->figure level function
sns.catplot(data=tips, x='day', y='total_bill', kind='strip', hue='sex')
```

```python
# swarmplot -> figure level function
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm')
```

```python
# swarmplot -> axes level function
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
# Box plot
sns.boxplot(data=tips, x='sex', y='total_bill')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
# using catplot
sns.catplot(data=tips, x='day', y='total_bill', kind='box')
```

```python
# hue
sns.boxplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
# single boxplot -> numerical col
sns.boxplot(data=tips, y='total_bill')
```

```python
# Violinplot = (Boxplot + KDEplot)
```

```python
# voilinplot
sns.violinplot(data=tips, x='day', y='total_bill')
```

```python
# using catplot
sns.catplot(data=tips, x='day', y='total_bill', kind='violin')
```

```python
# Hue
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex', split='True')
```

```python
# Barplot - By default calculate average
sns.barplot(data=tips, x='sex', y='total_bill', ci=None)
```

```python
# hue
sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker', ci=None )
```

```python
# estimator
sns.barplot(data=tips, x='sex', y='total_bill', hue='smoker', estimator=np.min)
```

```python
# point plot
sns.pointplot(data=tips, x='sex', y='total_bill', hue='smoker', estimator=np.min)
```

```python
# countplot
sns.countplot(data=tips, x='sex')
```

```python
# hue
sns.countplot(data=tips, x='sex', hue='day')
```

```python
# faceting using catplot
sns.catplot(data=tips, x='sex', y='total_bill', col='smoker', row='time', kind='box')
```

```python
Multi grid plots
 - Facet grid
 - pair grid
 - join grid
```

```python
# figure level -> relplot -> displot -> catplot -> lmplot
```

```python
# Facet grid
g = sns.FacetGrid(data=tips, col='day', row='time')
g.map(sns.violinplot, 'sex', 'total_bill')
g.add_legend()
```

```python
g = sns.FacetGrid(data=tips, col='day', row='time', hue='smoker')
g.map(sns.scatterplot,'tip','total_bill')
g.add_legend()
```

```python
# Pairplot
iris
```

```python
# pairplot
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

```python
# Pair Grid
g = sns.PairGrid(data=iris, hue='species')
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data=iris, hue='species', vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# Joinplot
sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
# JointGrid
g = sns.JointGrid(data=tips, x='total_bill', y='tip')
g.plot(sns.scatterplot,sns.histplot)
```
